```python
import pandas as pd
import matplotlib.pyplot as plt
data = {
    "Order_Hour": [10, 11, 1, 2, 4, 6, 7, 1],
    "Delivery_Time": [30, 45, 50, 25, 55, 35, 50, 25],
    "Delivery_Partner": ["Rahul", "Sneha", "Arjun", "Priya", "Kiran", "Meena", "Ravi", "Divya"],
    "Area": ["A", "B", "C", "A", "D", "B", "C", "D"],
    "Customer_Rating": [5, 4, 3, 5, 2, 4, 3, 5]}
df = pd.DataFrame(data)

avg_time = df["Delivery_Time"].mean()
print("Average Delivery Time:", avg_time)

peak_hour = df["Order_Hour"].mode()
print("Peak Order Hour:", peak_hour)

delay_area = df.groupby("Area")["Delivery_Time"].mean()
print("\nMost Delayed Areas:")
print(delay_area)

top_partner = df.groupby("Delivery_Partner")["Customer_Rating"].mean()
print("\nTop Rated Delivery Partners:")
print(top_partner)

delay_area.plot(kind="bar")
plt.title("Delivery Time by Area")
plt.xlabel("Area")
plt.ylabel("Minutes")
plt.show()

top_partner.plot(kind="bar")
plt.title("Delivery Partner Ratings")
plt.xlabel("Partner")
plt.ylabel("Rating")
plt.show()

print("\nPerformance Report")
print("Average delivery time is", avg_time, "minutes")
print("Peak order hour is", peak_hour)

print("\nSuggestions")
print("1. Increase staff during peak hours")
print("2. Improve delivery in delayed areas")
print("3. Train low-rated delivery partners")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    "Product": ["Phone", "Laptop", "Phone", "Headphones", "Phone"],
    "Purchases": [2, 1, 3, 1, 4],
    "Cart_Abandonment": [1, 0, 1, 0, 1],
    "Customer_Rating": [4, 5, 2, 3, 1],
    "Payment_Method": ["UPI", "Card", "UPI", "Cash", "UPI"]
}

df = pd.DataFrame(data)

print(df)
most_sold = df["Product"].mode()[0]
print("\nMost Sold Product:", most_sold)

high_revenue = df["Purchases"].max()
print("High Revenue Customer Purchases:", high_revenue)

payment = df["Payment_Method"].mode()[0]
print("Most Used Payment Method:", payment)

correlation = df["Purchases"].corr(df["Customer_Rating"])
print("Correlation:", correlation) #groupby gives average purchases for each product.

if correlation < 0:
    print("Products may have poor ratings")
else:
    print("Ratings are good")

product_count = df["Product"].value_counts()

plt.bar(product_count.index, product_count.values)
plt.title("Most Sold Products")
plt.xlabel("Products")
plt.ylabel("Count")
plt.show()
revenue = df.groupby("Product")["Purchases"].sum()
plt.pie(revenue, labels=revenue.index, autopct='%1.1f%%')
plt.title("High Revenue Products")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Simple Fraud Detection Data
data = {
    "Transaction_Amount": [500.50, 20000.75, 300.25, 45000.90, 700.40],
    "Location": ["India", "USA", "India", "Unknown", "India"],
    "Time": ["10AM", "2AM", "1PM", "3AM", "11AM"],
    "Account_Activity": [5, 50, 3, 80, 4]
}

df = pd.DataFrame(data)

print(df)

# Unusual Transactions
unusual = df[df["Transaction_Amount"] % 1 != 0]

print("\nUnusual Transactions:")
print(unusual)
# High Risk Accounts
high_risk = df[df["Account_Activity"] > 40]

print("\nHigh Risk Accounts:")
print(high_risk)

# Fraud Patterns
fraud = df[
    (df["Transaction_Amount"] > 10000) &
    (df["Time"].isin(["2AM", "3AM"]))
]

print("\nPossible Fraud Patterns:")
print(fraud)

# Graph
plt.bar(df.index, df["Transaction_Amount"])

plt.title("Transaction Amounts")
plt.xlabel("Transactions")
plt.ylabel("Amount")

plt.show()
```
